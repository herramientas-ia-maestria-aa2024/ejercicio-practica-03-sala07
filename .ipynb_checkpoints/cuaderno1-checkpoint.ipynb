```python
import random as rd
```

```python
def obtener_edad(d):
    edad = str(rd.randint(30,40))
    edad = edad + " años"
    return edad
```

```python
prov=['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabí']
```

```python
def obtener_provincia(p):
    provincia = rd.choice(prov)
    return provincia
```

```python
data['edad'] = data['nombre'].apply(obtener_edad)
```

```python
data['provincia'] = data['nombre'].apply(obtener_provincia)
```
